```python
import numpy as np
import pandas as pd
import numpy as np
import datetime
from datetime import date
from pandas.tseries.offsets import *
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
from matplotlib.dates import MonthLocator, YearLocator
import matplotlib.patches as mpatches
import matplotlib as matplotlib

import warnings
warnings.filterwarnings('ignore')
```

```python
today = date.today().strftime('%Y-%m-%d')
```

```python
import coinmetrics

cm = coinmetrics.Community()

begin_timestamp = "2010-07-18"
end_timestamp = "2021-05-04"

asset = "btc"
metric = "PriceUSD"
asset_data = cm.get_asset_data_for_time_range(asset, metric, begin_timestamp, end_timestamp)
```

```python
btc = coinmetrics.cm_to_pandas(asset_data)
btc.index = pd.to_datetime(btc.index)
btc.index = btc.index.tz_localize(None)
btc.columns = ['BTC']
btc = btc.dropna()
btc = btc[btc.BTC != '.']
btc['BTC'] = btc['BTC'].astype(float)
btc
```

```python
fig, ax1 = plt.subplots(1,1, figsize = (20, 10))

color = 'darkorange'
ax1.set_xlabel('Date')
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax1.set_ylabel('$', color=color)
ax1.plot(btc.index, btc['BTC'], color=color, label="BTC (linear)")
ax1.tick_params(axis='y', labelcolor=color)

color = 'gold'
ax2 = ax1.twinx()
ax2.plot(btc.index, btc['BTC'], color=color, label="BTC (log)")
ax2.tick_params(axis='y', labelcolor=color)

yloc = YearLocator(1, month=1, day=1)
ax1.xaxis.set_major_locator(yloc)
ax1.grid(True)
ax1.set_xlim([datetime.date(2010, 7, 18), datetime.date(2021, 5, 3)])

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper left")

ax2.set_yscale('log')

plt.title('BTC\n', fontsize=25)
plt.xticks(rotation=90)
plt.show()
```

```python
BTC_CYCLE1 = btc[(btc.index >= "2012-11-28") & (btc.index <= "2016-07-09")]
BTC_CYCLE1.rename(columns = {"BTC":"PRICE_CYCLE1"},inplace=True)
daysCount = pd.to_datetime(BTC_CYCLE1.index) - datetime.datetime(2012,11,28)
BTC_CYCLE1['DAYS_AFTER'] = daysCount
BTC_CYCLE1.set_index('DAYS_AFTER')
```

```python
BTC_CYCLE2 = btc[(btc.index >= "2016-07-10") & (btc.index <= "2020-05-11")]
BTC_CYCLE2.rename(columns = {"BTC":"PRICE_CYCLE2"},inplace=True)
daysCount = pd.to_datetime(BTC_CYCLE2.index) - datetime.datetime(2016,7,10)
BTC_CYCLE2['DAYS_AFTER'] = daysCount
BTC_CYCLE2.set_index('DAYS_AFTER')
```

```python
BTC_CYCLE3 = btc[(btc.index >= "2020-05-12") & (btc.index <= today)]
BTC_CYCLE3.rename(columns = {"BTC":"PRICE_CYCLE3"},inplace=True)
daysCount = pd.to_datetime(BTC_CYCLE3.index) - datetime.datetime(2020,5,12)
BTC_CYCLE3['DAYS_AFTER'] = daysCount
BTC_CYCLE3.set_index('DAYS_AFTER')
```

```python
BTC_CYCLES = pd.merge(BTC_CYCLE1, BTC_CYCLE2, how = 'outer', on = ['DAYS_AFTER'])
BTC_CYCLES = pd.merge(BTC_CYCLES, BTC_CYCLE3, how = 'outer', on = ['DAYS_AFTER'])
BTC_CYCLES.set_index('DAYS_AFTER')
```

```python
BTC_CYCLES['CYCLE1_FACTOR'] = BTC_CYCLES['PRICE_CYCLE1'] / 12.331660
BTC_CYCLES['CYCLE2_FACTOR'] = BTC_CYCLES['PRICE_CYCLE2'] / 649.959341
BTC_CYCLES['CYCLE3_FACTOR'] = BTC_CYCLES['PRICE_CYCLE3'] / 8817.250681
BTC_CYCLES
```

```python
fig, ax1 = plt.subplots(1,1, figsize = (20, 10))


ax1.set_xlabel('Days After Halving')
ax1.plot(BTC_CYCLES.index, BTC_CYCLES['CYCLE1_FACTOR'], color='darkorchid', label="Cycle 1 (2012)")
ax1.tick_params(axis='y', labelcolor='darkorchid')
plt.yscale('symlog')

ax2 = ax1.twinx()
ax2.plot(BTC_CYCLES.index, BTC_CYCLES['CYCLE2_FACTOR'], color='mediumseagreen', label="Cycle 2 (2016)")
ax2.tick_params(axis='y', labelcolor='mediumseagreen')
plt.yscale('symlog')

ax3 = ax1.twinx()
ax3.plot(BTC_CYCLES.index, BTC_CYCLES['CYCLE3_FACTOR'], color='orange', label="Cycle 3 (Current)")
plt.yscale('symlog')

ax1.set_ylim(0.75, 100)
ax2.set_ylim(0.75, 100)
ax3.set_ylim(0.75, 100)

ax1.grid(True)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax1.legend(lines + lines2 + lines3, labels + labels2 + labels3, loc="upper left")


plt.xticks(rotation=90)
plt.title('BTC Cycles\n', fontsize=25)
plt.show()
```

```python
BTC_REC1 = btc[(btc.index >= "2020-12-3") & (btc.index <= today)]
BTC_REC1.rename(columns = {"BTC":"TOP3"},inplace=True)
daysCount = pd.to_datetime(BTC_REC1.index) - datetime.datetime(2020,12,3)
BTC_REC1['DAYS_AFTER'] = daysCount
BTC_REC1.set_index('DAYS_AFTER')

```

```python
BTC_REC2 = btc[(btc.index >= "2017-02-23") & (btc.index <= "2020-12-15")]
BTC_REC2.rename(columns = {"BTC":"TOP2"},inplace=True)
daysCount = pd.to_datetime(BTC_REC2.index) - datetime.datetime(2017,2,23)
BTC_REC2['DAYS_AFTER'] = daysCount
BTC_REC2.set_index('DAYS_AFTER')
```

```python
BTC_REC3 = btc[(btc.index >= "2013-02-15") & (btc.index <= "2017-02-23")]
BTC_REC3.rename(columns = {"BTC":"TOP1"},inplace=True)
daysCount = pd.to_datetime(BTC_REC3.index) - datetime.datetime(2013,2,15)
BTC_REC3['DAYS_AFTER'] = daysCount
BTC_REC3.set_index('DAYS_AFTER')
```

```python
# BTC_RETOP = BTC_CYCLES
BTC_RETOP = pd.merge(BTC_REC3, BTC_REC2, how = 'outer', on = ['DAYS_AFTER'])
BTC_RETOP = pd.merge(BTC_RETOP, BTC_REC1, how = 'outer', on = ['DAYS_AFTER'])
BTC_RETOP.set_index('DAYS_AFTER')
```

```python
BTC_RETOP['TOP1_FACTOR'] = BTC_RETOP['TOP1'] / 27.171269
BTC_RETOP['TOP2_FACTOR'] = BTC_RETOP['TOP2'] / 1188.861786
BTC_RETOP['TOP3_FACTOR'] = BTC_RETOP['TOP3'] / 19470.311208
del BTC_RETOP['TOP1']
del BTC_RETOP['TOP2']
del BTC_RETOP['TOP3']
BTC_RETOP['fill'] = 0
pd.set_option('display.max_rows', BTC_RETOP.shape[0]+1)
print(BTC_RETOP)
```

```python
fig, ax1 = plt.subplots(1,1, figsize = (20, 10))

ax1.set_xlabel('Days Since Retopping')
ax1.set_ylabel('Return', color='black')
ax1.plot(BTC_RETOP.index, BTC_RETOP['TOP3_FACTOR'], color='orange', label="Cycle 3 (Current)")
ax1.tick_params(axis='y', labelcolor='darkorchid')
plt.yscale('symlog')

## set y ticks
# y_major = matplotlib.ticker.LogLocator(base = 10.0, numticks = 5)
# ax1.yaxis.set_major_locator(y_major)
# y_minor = matplotlib.ticker.LogLocator(base = 10.0, subs = np.arange(1.0, 10.0) * 0.1, numticks = 10)
# ax1.yaxis.set_minor_locator(y_minor)
# ax1.yaxis.set_minor_formatter(matplotlib.ticker.NullFormatter())

ax2 = ax1.twinx()
ax2.plot(BTC_RETOP.index, BTC_RETOP['TOP2_FACTOR'], color='mediumseagreen', label="Cycle 2 (2016)")
ax2.tick_params(axis='y', labelcolor='mediumseagreen')
plt.yscale('symlog')

ax3 = ax1.twinx()
ax3.plot(BTC_RETOP.index, BTC_RETOP['TOP1_FACTOR'], color='darkorchid', label="Cycle 1 (2012)")
ax1.tick_params(axis='y', labelcolor='orange')
plt.yscale('symlog')

ax1.set_ylim(-0.25, 50)
ax2.set_ylim(-0.25, 50)
ax3.set_ylim(-0.25, 50)

ax1.grid(True)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax1.legend(lines + lines2 + lines3, labels + labels2 + labels3, loc="upper left")

ax1.set_xlim([0, 600])

ax3.axes.get_yaxis().set_visible(False)
ax2.axes.get_yaxis().set_visible(False)


plt.xticks(rotation=90)
plt.title('BTC The Rehalving\n', fontsize=25)
pathToSave='no_zoom'
plt.savefig(pathToSave)
plt.show()
```
